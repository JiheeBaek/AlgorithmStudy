```python
N, M=list(map(int, input().split(' ')))
r, c, d=list(map(int, input().split(' ')))
matrix=[]
for i in range(N):
    matrix.append(list(map(int, input().split(' '))))
    
visited=[[False]*M for _ in range(N)]
answer=0
dx=[1, -1, 0, 0]
dy=[0, 0, 1, -1]

def turn_left(d): #방향 바꾸기
    if d==0: #북 -> 서
        d=3
    elif d==1: #동 -> 북
        d=0
    elif d==2: #남 -> 동
        d=1
    else: #서 ->남
        d=2
    return d

def next_cell(r, c, d): #한칸 전진했을 때의 위치
    if d==0:
        nr=r-1
        nc=c
    elif d==1:
        nr=r
        nc=c+1
    elif d==2:
        nr=r+1
        nc=c
    else:
        nr=r
        nc=c-1
    return nr, nc

def back_cell(r, c, d): #뒤로 후진했을 때의 위치
    if d==0:
        br=r+1
        bc=c
    elif d==1:
        br=r
        bc=c-1
    elif d==2:
        br=r-1
        bc=c
    else:
        br=r
        bc=c+1
    return br, bc

def check(r, c, d): #네 방향 모두 청소가 이미 되어있거나 벽인지 확인 
    for i in range(4):
        nr, nc=r+dx[i], c+dy[i]
        if visited[nr][nc]==True or matrix[nr][nc]==1:
            continue
        else:
            return False
    return True
            

while True:
    if matrix[r][c]==0 and visited[r][c]==False: # 1.현재 위치를 청소한다.
        visited[r][c]=True
        matrix[r][c]
        answer+=1
        
    if check(r, c, d): #2-c,d.네 방향 모두 청소가 이미 되어있거나 벽 -> True
        br, bc=back_cell(r, c, d)
        if matrix[br][bc]==1:  #뒤쪽 방향이 벽
            break              #작동을 멈춘다.
        else:
            r, c=br, bc #바라보는 방향을 유지한 채로 한 칸 후진
    else:
        td=turn_left(d) #왼쪽
        nr, nc=next_cell(r, c, td)
        if matrix[nr][nc]==0 and visited[nr][nc]==False: #청소하지 않은 공간 존재
            d=td #그 방향으로 회전
            r, c=nr, nc #한칸 전진 
        else:
            d=td #회전만
            
print(answer)
```
