```python
from collections import deque
from copy import deepcopy

N, M = map(int, input().split(' '))
office = []
for _ in range(N):
    office.append(list(map(int, input().split(' '))))


# 0을 가장 많이 바꾸는 경우로 선택
def cctv1(office, start):
    # 오른쪽, 왼쪽, 아래, 위
    dx = [0, 0, 1, -1]
    dy = [1, -1, 0, 0]
    count = []
    for i in range(4):
        cnt = 0
        queue = deque([start])
        while queue:
            x, y = queue.popleft()
            nx, ny = x + dx[i], y + dy[i]
            if nx >= 0 and nx < N and ny >= 0 and ny < M:
                if office[nx][ny] == 6:
                    break
                if office[nx][ny] == 0:
                    cnt += 1
                queue.append((nx, ny))
        count.append(cnt)

    d = count.index(max(count))
    queue = deque([start])
    while queue:
        x, y = queue.popleft()
        nx, ny = x + dx[d], y + dy[d]
        if nx >= 0 and nx < N and ny >= 0 and ny < M:
            if office[nx][ny] == 6:
                break
            if office[nx][ny] == 0:
                office[nx][ny] = -1
            queue.append((nx, ny))
    return office


def cctv2(office, start):
    count = []
    for i in range(2):
        if i==0:
            dx = [0, 0]  # horizontal
            dy = [-1, 1]
        else:
            dx=[-1, 1] #vertical
            dy=[0, 0]
        cnt=0
        for j in range(2):
            queue=deque([start])
            while queue:
                x, y = queue.popleft()
                nx, ny = x + dx[j], y + dy[j]
                if nx >= 0 and nx < N and ny >= 0 and ny < M:
                    if office[nx][ny] == 6:
                        break
                    if office[nx][ny] == 0:
                        cnt += 1
                    queue.append((nx, ny))
        count.append(cnt)
    d = count.index(max(count))
    if d==0:
        dx = [0, 0]  # horizontal
        dy = [-1, 1]
    else:
        dx=[-1, 1] #vertical
        dy=[0, 0]
    for j in range(2):
        queue=deque([start])
        while queue:
            x, y = queue.popleft()
            nx, ny = x + dx[j], y + dy[j]
            if nx >= 0 and nx < N and ny >= 0 and ny < M:
                if office[nx][ny] == 6:
                    break
                if office[nx][ny] == 0:
                    office[nx][ny]=-1
                queue.append((nx, ny))
    return office


def cctv3(office, start):
    count = []
    for i in range(4):
        if i == 0: #위, 오른쪽
            dx = [-1, 0]
            dy = [0, 1]
        elif i == 1: #오른쪽, 아래
            dx = [0, 1]
            dy = [1, 0]
        elif i == 2: #아래, 왼쪽
            dx = [0, 1]
            dy = [-1, 0]
        elif i == 3: #왼쪽, 위
            dx = [0, -1]
            dy = [-1, 0]
        cnt=0
        for j in range(2):
            queue=deque([start])
            while queue:
                x, y = queue.popleft()
                nx, ny = x + dx[j], y + dy[j]
                if nx >= 0 and nx < N and ny >= 0 and ny < M:
                    if office[nx][ny] == 6:
                        break
                    if office[nx][ny] == 0:
                        cnt += 1
                    queue.append((nx, ny))
        count.append(cnt)
    d = count.index(max(count))
    if d == 0: #위, 오른쪽
        dx = [-1, 0]
        dy = [0, 1]
    elif d == 1: #오른쪽, 아래
        dx = [0, 1]
        dy = [1, 0]
    elif d == 2: #아래, 왼쪽
        dx = [0, 1]
        dy = [-1, 0]
    elif d == 3: #왼쪽, 위
        dx = [0, -1]
        dy = [-1, 0]
        
    for j in range(2):
        queue=deque([start])
        while queue:
            x, y = queue.popleft()
            nx, ny = x + dx[j], y + dy[j]
            if nx >= 0 and nx < N and ny >= 0 and ny < M:
                if office[nx][ny] == 6:
                    break
                if office[nx][ny] == 0:
                    office[nx][ny]=-1
                queue.append((nx, ny))
    return office

def cctv4(office, start):
        count = []
    for i in range(4):
        if i == 0:  # 왼쪽, 위, 오른쪽
            dx = [0, -1, 0]
            dy = [-1, 0, 1]
        elif i == 1:  # 위, 오른쪽, 아래
            dx = [-1, 0, 1]
            dy = [0, 1, 0]
        elif i == 2:  # 왼쪽, 아래, 오른쪽
            dx = [0, 1, 0]
            dy = [-1, 0, 1]
        elif i == 3:  # 아래, 왼쪽, 위
            dx = [1, 0, -1]
            dy = [0, -1, 0]
        cnt=0
        for j in range(3):
            queue=deque([start])
            while queue:
                x, y = queue.popleft()
                nx, ny = x + dx[j], y + dy[j]
                if nx >= 0 and nx < N and ny >= 0 and ny < M:
                    if office[nx][ny] == 6:
                        break
                    if office[nx][ny] == 0:
                        cnt += 1
                    queue.append((nx, ny))
        count.append(cnt)
    d = count.index(max(count))
    if d == 0:  # 왼쪽, 위, 오른쪽
        dx = [0, -1, 0]
        dy = [-1, 0, 1]
    elif d == 1:  # 위, 오른쪽, 아래
        dx = [-1, 0, 1]
        dy = [0, 1, 0]
    elif d == 2:  # 왼쪽, 아래, 오른쪽
        dx = [0, 1, 0]
        dy = [-1, 0, 1]
    elif d == 3:  # 아래, 왼쪽, 위
        dx = [1, 0, -1]
        dy = [0, -1, 0]
        
    for j in range(3):
        queue=deque([start])
        while queue:
            x, y = queue.popleft()
            nx, ny = x + dx[j], y + dy[j]
            if nx >= 0 and nx < N and ny >= 0 and ny < M:
                if office[nx][ny] == 6:
                    break
                if office[nx][ny] == 0:
                    office[nx][ny]=-1
                queue.append((nx, ny))
    return office

def cctv5(office, start):
    dx = [0, 0, 1, -1]
    dy = [1, -1, 0, 0]
    for i in range(4):
        queue=deque([start])
        while queue:
            x, y = queue.popleft()
            nx, ny = x + dx[i], y + dy[i]
            if nx >= 0 and nx < N and ny >= 0 and ny < M:
                if office[nx][ny] == 6:
                    break
                if office[nx][ny] == 0:
                    office[nx][ny] = -1
                queue.append((nx, ny))
    return office

for i in range(N):
    for j in range(M):
        if office[i][j] == 1:
            office = cctv1(office, (i, j))
        elif office[i][j] == 2:
            office = cctv2(office, (i, j))
        elif office[i][j] == 3:
            office = cctv3(office, (i, j))
        elif office[i][j] == 4:
            office = cctv4(office, (i, j))
        elif office[i][j] == 5:
            office = cctv5(office, (i, j))
        else:
            continue
result = 0
for i in range(N):
    print(office[i])
    result += office[i].count(0)

print(result)
```
